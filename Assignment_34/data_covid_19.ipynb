```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
data_covid = pd.read_csv('covid_19.csv')
data_covid = data_covid.drop('idx', axis=1)
data_covid.head()
```

```python
data_covid_Monthly = data_covid.query("'2020-04-01' <= Date <= '2020-04-31'").groupby(['CountryRegion'])['New_cases'].sum().reset_index()
data_covid_Monthly.sort_values('New_cases',ascending= False)
data_covid_Monthly_8Country = data_covid_Monthly.sort_values('New_cases',ascending= False)[0:8].reset_index()
data_covid_Monthly_8Country = data_covid_Monthly_8Country.drop('index',axis=1)
print(data_covid_Monthly_8Country)
```

```python
#data_covid_iran_Dayily_202105 = data_covid.query(("CountryRegion == 'Iran' and '2021-05-01' <= Date <= '2021-05-31'"))
data_covid_iran_Dayily_2021 = data_covid.query(("CountryRegion == 'Iran' and '2021-01-01' <= Date <= '2021-12-30'"))
data_covid_iran_Dayily_2021
```

```python
data_covid_iran_Dayily_2021['Date'] = pd.to_datetime(data_covid_iran_Dayily_2021['Date'])
data_covid_iran_Dayily_2021['Month'] = data_covid_iran_Dayily_2021['Date'].map(lambda date: date.month ) 
data_covid_iran_Dayily_2021
```

```python

plt.figure(figsize= (12,6))
selectM = 7
data_covid_iran_Dayily_2021[data_covid_iran_Dayily_2021["Month"]== selectM].reset_index()[['New_cases','New_deaths']].plot(kind= 'area',
legend= True,
stacked = False)
stitle = "Covid-19 in Iran - 2021-"+str(selectM)
plt.suptitle(stitle)
plt.xlim(0,30)
plt.margins(0.02)
# plt.xticks(np.arange(0,30, step=1))
plt.grid(alpha = 0.5)


```
