```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv
from scipy import stats
```

```python
X = np.random.uniform(0,10, 200)
Y = X  * np.random.uniform(1, 3, 200) + np.random.uniform(-4, 4, 200)
```

```python
idx = np.where(Y<=20)
Y = Y[Y <= 20]
X = X[idx]
```

```python
Y.shape
X.shape
```

```python
X = X.reshape(X.shape[0], 1)
Y = Y.reshape(Y.shape[0], 1)
```

```python
# random fit line slope

plt.figure(figsize=[12,8])
plt.subplot(2,2,1)
m = 1.2
y_pred = m * X
plt.scatter(X, Y, c='k', s= 3, label='Data')
plt.plot(X, y_pred, c= 'r', lw= 4, label='Random_line')
plt.legend()

plt.subplot(2,2,2)
m = 0.5
y_pred = m * X
plt.scatter(X, Y, c='k', s= 3, label='Data')
plt.plot(X, y_pred, c= 'r', lw= 4, label='Random_line')
plt.legend()

plt.subplot(2,2,3)
m = 2.1
y_pred = m * X
plt.scatter(X, Y, c='k', s= 3, label='Data')
plt.plot(X, y_pred, c= 'r', lw= 4, label='Random_line')
plt.legend()

plt.subplot(2,2,4)
m = 1.75
y_pred = m * X
plt.scatter(X, Y, c='k', s= 3, label='Data')
plt.plot(X, y_pred, c= 'r', lw= 4, label='Random_line')
plt.legend()
```

```python
lls = np.matmul( inv(np.matmul(X.T, X)) , np.matmul( X.T, Y) ) 
lls = lls[0,0]
lls
```

```python
y_pred = lls * X
```

```python
plt.scatter(X, Y, c='k', s= 3, label='Data')
plt.plot(X, y_pred, c='r', lw=4, label='Predict_line')
plt.legend()
```

```python
x_test = 10.5
answer = x_test * lls
answer
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(X[:,0], Y[:,0])
slope
```

```python
y_pred2 = intercept + slope * X
```

```python
plt.scatter(X, Y, c='k', s= 3, label='Data')
plt.plot(X, y_pred, c='r', lw=4, label='Predict_Line')
plt.plot(X, y_pred2, c='g', lw=4, label = 'Scipy_predict_Line')
plt.legend()
```
