```python
# marker
# You can use the keyword argument markersize or the shorter version, ms to set the size of the markers
# You can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers
# You can use the keyword argument markerfacecolor or the shorter mfc to set the color inside the edge of the markers
# Use both the mec and mfc arguments to color of the entire marker
# You can also use Hexadecimal color values
# Or any of the 140 supported color names : https://www.w3schools.com/colors/colors_names.asp

ypoints = np.array([3, 8, 1, 5])
flag = True
while flag:
    type_marker = input('type point Select between: o, *, ., ,, x, X, +, P, s, D, d, p, H, h, v, ^, <, >, 1, 2, 3, 4, |, _ :  ')
    type_line = input('type line Select between: -, :, --, -. :  ')
    type_color = input('type color Select between: r, g, b, c, m, y, k, w :  ')
    marker_size = input('marker size: ')
    marker_edge_color = input('Enter edge color of marker (Ex: #A52A2A) : ')
    marker_inside_color = input('Enter face color of marker (Ex: #FF7F50): ')
    marker_type_select = (type_marker+ type_line +  type_color)
    plt.plot(ypoints,  marker_type_select, ms = marker_size, mec = marker_edge_color, mfc = marker_inside_color  )
    plt.show()
        
    user_input = input('try again?(y/n): ')
    if user_input == 'n':
        flag = False

```

```python
# ___________________________________Matplotlib line _______________________________________________
# linestyle
# You can use the keyword argument linestyle, or shorter ls, to change the style of the plotted line
# Shorter Syntax : The line style can be written in a shorter syntax: 
# linestyle can be written as ls. dotted can be written as :.  dashed can be written as --.
# You can choose any of these styles: 
#   Style                Or  :
#  'solid' (default)	 '-'	
#  'dotted'	             ':'	
#  'dashed'	             '--'	
#  'dashdot'	         '-.'	
#  'None'	             '' or ' '

# Line Color
# You can use the keyword argument color or the shorter c to set the color of the line
# You can also use Hexadecimal color values
# Or any of the 140 supported color names : https://www.w3schools.com/colors/colors_names.asp

# Line Width
# You can use the keyword argument linewidth or the shorter lw to change the width of the line.
# The value is a floating number, in points
import matplotlib.pylab as plt
import numpy as np

ypoints = np.array([3, 8, 1, 5])
flag = True
while flag:
    type_line_Style = input('type line Select between: -, :, --, -., ''or'', solid, dotted, dashed, dashdot, None :  ')
    type_line_color = input('Enter line color of marker (Ex: #FF7F50) : ')
    type_line_width = float(input('Enter line width: '))
    plt.plot(ypoints,  ls = type_line_Style , c = type_line_color, lw = type_line_width )
    plt.show()
    
    user_input = input('try again?(y/n): ')
    if user_input == 'n':
        flag = False

```

```python
# ___________________________________Matplotlib Labels _______________________________________________
# Create Labels for a Plot
# With Pyplot, you can use the xlabel() and ylabel() functions to set a label for the x- and y-axis.
# With Pyplot, you can use the title() function to set a title for the plot.
# Set Font Properties for Title and Labels
# You can use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels.
# Position the Title
# You can use the loc parameter in title() to position the title.
# Legal values are: 'left', 'right', and 'center'. Default value is 'center'.
import matplotlib.pylab as plt
import numpy as np

xpoints = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
ypoints = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

flag = True
while flag:
    figure_x_Label = input('Enter x axis Label :  ')
    figure_y_Label = input('Enter y axis label : ')
    figure_title = input('Enter figure title : ')
    figure_title_location = input('Enter figure title Location between: right,left,center :  ')
    font_properties = ['1','2','3']
    for i in range(3):
        font_family = input('Enter font family : ')
        font_color = input('Enter font color : ')
        font_size = int(input('Enter font size : '))
        font_properties[i] = {'family' : font_family, 'color' : font_color, 'size': font_size}
            
    plt.xlabel(figure_x_Label, font_properties[0])
    plt.ylabel(figure_y_Label, font_properties[1])
    plt.title(figure_title, font_properties[2], loc = figure_title_location)
    plt.plot(xpoints, ypoints)
    plt.show()
    
    user_input = input('try again?(y/n): ')
    if user_input == 'n':
        flag = False

```

```python
# Add Grid Lines to a Plot
# With Pyplot, you can use the grid() function to add grid lines to the plot.
# Specify Which Grid Lines to Display
# You can use the axis parameter in the grid() function to specify which grid lines to display.
# Legal values are: 'x', 'y', and 'both'. Default value is 'both'.
# Set Line Properties for the Grid
# You can also set the line properties of the grid, like this: grid(color = 'color', linestyle = 'linestyle', linewidth = number).

xpoints = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
ypoints = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

flag = True
while flag:
    grid_lines_display = input('Enter grid lines display between: x, y, both :  ')
    type_line_Style = input('type line Select between: -, :, --, -., ''or'', solid, dotted, dashed, dashdot, None :  ')
    type_line_color = input('Enter line color of marker (Ex: #FF7F50) : ')
    type_line_width = float(input('Enter line width: '))
    
    plt.xlabel('Average Pulse')
    plt.ylabel('Calorie Burnage')
    plt.title('Sports Watch Data')
    plt.plot(xpoints, ypoints)
    plt.grid(axis = grid_lines_display, color = type_line_color, linestyle = type_line_Style, linewidth = type_line_width)
    plt.show()
   
    
    user_input = input('try again?(y/n): ')
    if user_input == 'n':
        flag = False
```

```python
# ___________________________________Matplotlib subplot _______________________________________________
# Display Multiple Plots
# With the subplots() function you can draw multiple plots in one figure
# Title
# You can add a title to each plot with the title() function:
# Super Title
# You can add a title to the entire figure with the suptitle() function
import matplotlib.pylab as plt
import numpy as np
import random

flag = True
while flag:

    num_rows_subplots = int(input("Enter number of rows : "))
    num_cols_subplots = int(input("Enter number of cols : "))
    sum = num_rows_subplots * num_cols_subplots
    ypoints = [[]] * sum

    for i in range(0, sum):
        
        ypoints[i] = random.sample(range(1, 30), 5)
        title_subplot = input('Enter subplot title : ')
        plt.subplot(num_rows_subplots, num_cols_subplots, i+1)
        plt.plot( ypoints[i])
        plt.title(title_subplot)

    super_title = input('Enter super title: ')
    plt.suptitle(super_title)
    plt.show()
    
    user_input = input('try again?(y/n): ')
    if user_input == 'n':
        flag = False
```

```python
# ___________________________________Matplotlib scatter plot _______________________________________________
# Creating Scatter Plots
# With Pyplot, you can use the scatter() function to draw a scatter plot.
# The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis
import matplotlib.pylab as plt
import numpy as np

xpoints = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
ypoints = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

#plt.subplot(3,3,1)
plt.scatter(xpoints, ypoints)
plt.title('scatter plot')
plt.show()
```

```python
# Compare Plots
# In the example above, there seems to be a relationship between speed and age, but what if we plot the observations from another day as well? Will the scatter plot tell us something else?

#day one, the age and speed of 13 cars:
xpoints1 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
ypoints1 = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# plt.subplot(3,3,2)
plt.scatter(xpoints1, ypoints1)

#day two, the age and speed of 15 cars:
xpoints2 = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
ypoints2 = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(xpoints2, ypoints2)
plt.title('compare scatter plots')
plt.show()
```

```python
# Colors
# You can set your own color for each scatter plot with the color or the c argument

# plt.subplot(3,3,3)

plt.scatter(xpoints1, ypoints1, color = 'hotpink')
plt.scatter(xpoints2, ypoints2, color = '#88c999')
plt.title('color for each scatter plot')
plt.show()
```

```python
# Color Each Dot
# You can even set a specific color for each dot by using an array of colors as value for the c argument

colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

# plt.subplot(3,3,4)
plt.scatter(xpoints1, ypoints1, c = colors)
plt.title('color for each dot')
plt.show()
```

```python
# ColorMap
# The Matplotlib module has a number of available colormaps.
# A colormap is like a list of colors, where each color has a value that ranges from 0 to 100.
# Create a color array, and specify a colormap in the scatter plot
# How to Use the ColorMap
# You can specify the colormap with the keyword argument cmap with the value of the colormap, in this case 'viridis' which is one of the built-in colormaps available in Matplotlib.
# In addition you have to create an array with values (from 0 to 100), one value for each of the point in the scatter plot

colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

# plt.subplot(3,3,5)
plt.scatter(xpoints1, ypoints1, c= colors, cmap='viridis')
plt.title('color map')
plt.show()
```

```python
# You can include the colormap in the drawing by including the plt.colorbar() statement
# Available ColorMaps 
# You can choose any of the built-in colormaps : https://www.w3schools.com/python/matplotlib_scatter.asp

# plt.subplot(3,3,6)
plt.scatter(xpoints1, ypoints1, c= colors, cmap='viridis')
plt.colorbar()
plt.title('color bar')
plt.show()
```

```python
# Size
# You can change the size of the dots with the s argument.
# Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis:

sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

# plt.subplot(3,3,7)

plt.scatter(xpoints1, ypoints1, s=sizes)
plt.title('size dots')
plt.show()
```

```python
# Alpha
# You can adjust the transparency of the dots with the alpha argument.
# Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis:

# plt.subplot(3,3,8)

plt.scatter(xpoints1, ypoints1, s=sizes, alpha=0.5)
plt.title('transparency of the dots')
plt.show()
```

```python
# Combine Color Size and Alpha
# You can combine a colormap with different sizes on the dots. This is best visualized if the dots are transparent
# Create random arrays with 100 values for x-points, y-points, colors and sizes:

xpoints = np.random.randint(100, size=(100))
ypoints = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

#plt.subplot(3,3,9)

plt.scatter(xpoints, ypoints, c= colors, s= sizes, cmap='nipy_spectral', alpha=0.5 )
plt.title('combine color, size and alpha')
plt.show()
```

```python
# ___________________________________Matplotlib Bars _______________________________________________
# Creating Bars
# With Pyplot, you can use the bar() function to draw bar graphs
import matplotlib.pylab as plt
import numpy as np

xpoints = np.array(["A", "B", "C", "D"])
ypoints = np.array([3, 8, 1, 10])

flag = True
while flag:
    
    horizontal_bar = input('do you want horizontal_bar? (y/n) : ')
    color_bar = input('Enter color of bar (Ex: #FF7F50) : ')
    
    
    if horizontal_bar == 'y':
        height_bar = float(input('Enter bar height (default is 0.8): '))
        plt.barh(xpoints, ypoints, color= color_bar, height= height_bar)
    else:
        width_bar = float(input('Enter bar width (default is 0.8): '))
        plt.bar(xpoints, ypoints, color= color_bar, width= width_bar)

    plt.title('Matplotlib Bars')
    plt.show()
        
    user_input = input('try again?(y/n): ')
    if user_input == 'n':
        flag = False
```

```python
# ___________________________________Matplotlib pie charts _______________________________________________
# Creating Pie Charts
# With Pyplot, you can use the pie() function to draw pie charts
# Labels
# Add labels to the pie chart with the label parameter.
# The label parameter must be an array with one label for each wedge
# The startangle parameter is defined with an angle in degrees, default angle is 0
# Explode
# Maybe you want one of the wedges to stand out? The explode parameter allows you to do that.
# The explode parameter, if specified, and not None, must be an array with one value for each wedge.
# Each value represents how far from the center each wedge is displayed
# Shadow
# Add a shadow to the pie chart by setting the shadows parameter to True
import matplotlib.pylab as plt
import numpy as np

xpoints = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

flag = True
while flag:
    pie_start_angel = float(input("Enter start angle degree (default is 0) : "))
    pie_shadow = input('do you want shadow? (y/n) : ')
    pie_explode = [[]] * 4
    wedges_color = [[]] * 4
    pie_colors = input('do you want change wedges color? (y/n) : ')
    if pie_colors == 'y':
        for i in range(4):
            wedges_color[i] = input('Enter color: ')
    else:
        wedges_color = None        
    for i in range(4):
        pie_explode[i] = float(input("Enter one value for each wedge explode (default is 0) : "))
        
    if pie_shadow == 'y':
        plt.pie(xpoints, labels= mylabels, startangle= pie_start_angel, explode= pie_explode, colors=wedges_color, shadow=True) 
    else:
        plt.pie(xpoints, labels= mylabels, startangle= pie_start_angel, explode= pie_explode, colors=wedges_color)

    plt.title('Pie Charts')
    plt.legend(title = "Four Fruits:")
    plt.show()
    
    
    user_input = input('try again?(y/n): ')
    if user_input == 'n':
        flag = False
```
